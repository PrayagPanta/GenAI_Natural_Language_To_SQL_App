The departure time above is not very much informative so we are going to change it in the datetime format so that we get a better idea of the time.

```python
# Creating a function to change the way of representation of time in the column
def Format_Hourmin(hours):
        if hours == 2400:
            hours = 0
        else:
            hours = "{0:04d}".format(int(hours))
            Hourmin = datetime.time(int(hours[0:2]), int(hours[2:4]))
            return Hourmin
```

```python
data2['Actual_Departure'] =data1['DEPARTURE_TIME'].apply(Format_Hourmin)
data2['Actual_Departure']
```

CREATE THE Cleaned_flights.db in chunks:

```python
import sqlite3
import pandas as pd

# Assume Flights is already loaded with your data

# Establish SQLite connection
conn = sqlite3.connect('cleaned_flights.db')

# Create the table using the schema of the Flights DataFrame
Flights.head(0).to_sql('Flight', conn, if_exists='replace', index=False)

# Define the chunk size
chunksize = 1000

# Save the DataFrame in chunks
for start in range(0, len(Flights), chunksize):
    end = start + chunksize
    chunk_df = Flights.iloc[start:end]
    chunk_df.to_sql('Flight', conn, if_exists='append', index=False)

# Close the connection
conn.close()
```
